Sometimes we need to store more than one value in a variable. In this case, we can use these types of variables
<li>list
<li>tuple
<li>set
<li>dictionary
    
### List
<li> allows duplicates
<li> is changeable
<li> values are ordered
    

### Tuple
<li> allows duplicates
<li> is unchangeable
<li> values are ordered
    
### Set
<li> does not allow duplicates
<li> is unchangeable
<li> values are unordered
    
    
### Dictionary
<li> does not allow duplicates
<li> is changeable
<li> values are ordered
<li> works with key:value pairs
</p><code>
list_example = ['Prague','Brno','Ostrava']
tuple_example = ('Prague','Brno','Ostrava')
set_example = {'Prague','Brno','Ostrava'}
dictionary_example = {'Prague':'capital city','Brno':'county seat','Ostrava':'county seat'}
</code>

You can access the whole object / collection of values or you can reach a specific item by its position / name.
For example

<code>
list_example[0] #returns 'Prague'
tuple_example[0] #returns 'Prague'
set_example[0] #returns an error because sets are not ordered
dictionary_example['Prague'] #returns 'capital city'
</code>

### Iteration
Or, you can go trough the object / collection item by item. This process is called iteration and is accessible via keyword for.
And example would be

<code>
#we store an item value into a variable item, then we continue to another step
for item in list_example: 
    #this nested step is intended, this is how we tell Python that it is a different block of code
    print(item)
</code>

In case we iterate over a dictionary, we can access:
<li> only a key
<li> only a value
<li> both a key and its value

both key and value by assigning two variables, for example

<code>

for k in dictionary_example: #access only keys    
    print()


for k in dictionary_example.keys(): #access only keys  
    print(k)
    
for v in dictionary_example.values(): #access only values  
    print(v)


for k,v in dictionary_example.items(): #access keys and values  
    print(k,v)
</code>
    
    
### Add or remove values into objects

#### List

##### Add
<li> list_example.append('Plzen')
<li> list_example.insert(1,'Plzen')

##### Remove
<li> list_example.remove('Plzen')
<li> list_example.pop(1)
    
    
#### Sets

##### Add
<li> set_example.add('Plzen')

##### Remove
<li> set_example.remove('Plzen')
<li> set_example.pop()
    
#### Dictionary

##### Add
<li> dictionary_example['Plzen] = 'county seat'
<li> dictionary_example.update({'Plzen':'county seat'})

##### Remove
<li> dictionary_example.pop('Plzen')
<li> dictionary_example.popitem()

Since we store value in variables, it might be handy to know how to check these values against a condition. For example:
- if a variable value equals / does not / is greater than / is less than equal another value
- if there is a specific item in object

In Python, these logical conditions are supported

<li> equals: a == b
<li> not equals: a != b
<li> less than: a < b
<li> less or equal to: a <= b
<li> greater than: a > b
<li> greater or equal to: a >= b
    
For example

<code>
if a > b:
    print('a is greater')
    
</code>
    
Other than that, we can check multiple conditions or define, what should happen in case "if" returns False (the condition is not true)

<code>
if a > b:
    print('a is greater')
elif a == b:
    print('they equal')
else:
    print('a is not greater and they do not equal)

</code>

Multiple conditions testing
We can test whether all or at least one conditions are met by using these keywords:
<li> and
<li> or
    
 
    

### Loops
You've alredy seen for loop while learning about lists, tuples etc. Lets dive more deeper into loops now.
Python has two primitive loops commands:
<li>while 
<li>for

### While loop
The while looop excecutes a statement as long as the condition is true

<code>
i = 0
while i < 6:
    print(f'{i} is less than 6')
    i +=1
</code>
    
### For loop
For loop is used for iterating over a sequence whether it is string, list, tuple, dictionary etc.
If not set up differently, for loops over each item in a sequence.
If needed, we can tell for loop to change this default behavior by commands:
<li> break : stop iteration if condition is met
<li> continue: skip iteration if condition is met
    

<code>
dictionary_example = {'Prague':'capital city','Brno':'county seat','Ostrava':'county seat'}
for k,v in dictionary.items():
    if v == 'capital city':
        print(f'{k} is a capital city')
    else:
        continue
</code>

<code>
dictionary_example = {'Prague':'capital city','Brno':'county seat','Ostrava':'county seat'}
for k,v in dictionary.items():
    if v == 'capital city':
        print(f'{k} is a capital city')
    else:
        break
</code>
